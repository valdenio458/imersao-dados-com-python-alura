# Aula 1 - Análise de Dados com Pandas


```python
import pandas as pd
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/guilhermeonrails/data-jobs/refs/heads/main/salaries.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.shape
```

```python
linhas, colunas = df.shape[0], df.shape[1]
print(f"O DataFrame possui {linhas} linhas e {colunas} colunas.")
print("Linhas:", linhas)
print("Colunas:", colunas)
```

```python
df.columns
```

```python
colunas_traduzidas = {
    'work_year': 'ano',
    'experience_level': 'senioridade',
    'employment_type': 'contrato',
    'job_title': 'cargo',
    'salary': 'salario',
    'salary_currency': 'moeda',
    'salary_in_usd': 'usd',
    'employee_residence': 'residencia',
    'remote_ratio': 'remoto',
    'company_location': 'empresa',
    'company_size': 'tamanho_empresa'
}
df.rename(columns=colunas_traduzidas, inplace=True)
print(df.columns)
```

```python
df['senioridade'].value_counts()
```

```python
df['senioridade'] = df['senioridade'].replace({
    'SE': 'Sênior',
    'MI': 'Pleno',
    'EN': 'Junior',
    'EX': 'Executivo'
})
df['senioridade'].value_counts()
```

```python
df['contrato'].value_counts()
```

```python
df['contrato'] = df['contrato'].replace({
    'FT': 'Tempo Integral',
    'CT': 'Contrato',
    'PT': 'Tempo Parcial',
    'FL': 'Freelancer'
})
df['contrato'].value_counts()
```

```python
df['tamanho_empresa'].value_counts()
```

```python
df['tamanho_empresa'] = df['tamanho_empresa'].replace({
    'S': 'Pequena',
    'M': 'Média',
    'L': 'Grande'
})
df['tamanho_empresa'].value_counts()
```

```python
df['remoto'].value_counts()
```

```python
df['remoto'] = df['remoto'].replace({
    0: 'Presencial',
    50: 'Híbrido',
    100: 'Remoto'
})
df['remoto'].value_counts()
```

```python
df.head()
```

```python
df.describe(include='object')
```

```python
df.describe()
```
