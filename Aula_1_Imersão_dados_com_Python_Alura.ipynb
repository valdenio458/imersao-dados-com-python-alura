# Aula 1 - Análise de Dados com Pandas


```python
import pandas as pd
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/guilhermeonrails/data-jobs/refs/heads/main/salaries.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.shape
```

```python
linhas, colunas = df.shape[0], df.shape[1]
print(f"O DataFrame possui {linhas} linhas e {colunas} colunas.")
print("Linhas:", linhas)
print("Colunas:", colunas)
```

```python
df.columns
```

```python
colunas_traduzidas = {
    'work_year': 'ano',
    'experience_level': 'senioridade',
    'employment_type': 'contrato',
    'job_title': 'cargo',
    'salary': 'salario',
    'salary_currency': 'moeda',
    'salary_in_usd': 'usd',
    'employee_residence': 'residencia',
    'remote_ratio': 'remoto',
    'company_location': 'empresa',
    'company_size': 'tamanho_empresa'
}
df.rename(columns=colunas_traduzidas, inplace=True)
print(df.columns)
```

```python
df['senioridade'].value_counts()
```

```python
df['senioridade'] = df['senioridade'].replace({
    'SE': 'Sênior',
    'MI': 'Pleno',
    'EN': 'Junior',
    'EX': 'Executivo'
})
df['senioridade'].value_counts()
```

```python
df['contrato'].value_counts()
```

```python
df['contrato'] = df['contrato'].replace({
    'FT': 'Tempo Integral',
    'CT': 'Contrato',
    'PT': 'Tempo Parcial',
    'FL': 'Freelancer'
})
df['contrato'].value_counts()
```

```python
df['tamanho_empresa'].value_counts()
```

```python
df['tamanho_empresa'] = df['tamanho_empresa'].replace({
    'S': 'Pequena',
    'M': 'Média',
    'L': 'Grande'
})
df['tamanho_empresa'].value_counts()
```

```python
df['remoto'].value_counts()
```

```python
df['remoto'] = df['remoto'].replace({
    0: 'Presencial',
    50: 'Híbrido',
    100: 'Remoto'
})
df['remoto'].value_counts()
```

```python
df.head()
```

```python
df.describe(include='object')
```

```python
df.describe()
```

# Aula 2 - Preparação e Limpeza dos Dados


```python
df.isnull().sum()
```

```python
df['ano'].unique()
```

```python
# Exibindo os valores nulos
df[df.isnull().any(axis=1)]
```

```python
df_limpo = df.dropna()
```

```python
df_limpo.isnull().sum()
```

```python
df_limpo.head()
```

```python
df_limpo.info()
```

```python
df_limpo = df_limpo.assign(ano = df_limpo['ano'].astype(int))
df_limpo
```

# Aula 3 - Visualização de Dados


```python
df_limpo.head()
```

```python
df_limpo['senioridade'].value_counts().plot(kind='bar',title="Distribuição de Senioridade")
```

```python
import seaborn as sns
```

```python
sns.barplot(data=df_limpo, x='senioridade', y='usd')
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(8, 5))
sns.barplot(data=df_limpo, x='senioridade', y='usd')
plt.title("Salário Médio por Nível de Seniorioridade")
plt.xlabel("Seniorioridade")
plt.ylabel("Salário Médio (USD)")
plt.show()
```

```python
df_limpo.groupby('senioridade')['usd'].mean().sort_values(ascending=False)
```

```python
ordem = df_limpo.groupby('senioridade')['usd'].mean().sort_values(ascending=False).index
ordem
```

```python
plt.figure(figsize=(8, 5))
sns.barplot(data=df_limpo, x='senioridade', y='usd', order=ordem)
plt.title("Salário Médio por Nível de Seniorioridade")
plt.xlabel("Seniorioridade")
plt.ylabel("Salário Médio (USD)")
plt.show()
```

```python
plt.figure(figsize=(8, 4))
sns.histplot(df_limpo['usd'], bins = 50, kde=True)
plt.title("Distribuição do Salários Anuais")
plt.xlabel("Salário em USD")
plt.ylabel("Frequência")
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.boxplot(x=df_limpo['usd'])
plt.title("Boxplot Salários Anuais")
plt.xlabel("Salário em USD")
plt.show()
```

```python
ordem_senioridade = ['Junior', 'Pleno', 'Sênior', 'Executivo']

plt.figure(figsize=(8, 5))
sns.boxplot(x='senioridade', y='usd', data=df_limpo, order=ordem_senioridade)
plt.title("Distribuição por senioridade")
plt.xlabel("Salário em USD")
plt.show()
```

```python
ordem_senioridade = ['Junior', 'Pleno', 'Sênior', 'Executivo']

plt.figure(figsize=(8, 5))
sns.boxplot(x='senioridade', y='usd', data=df_limpo, order=ordem_senioridade, palette= 'Set2', hue= 'senioridade' )
plt.title("Distribuição por senioridade")
plt.xlabel("Salário em USD")
plt.show()
```

```python
import plotly.express as px
```

```python
senioridade_media_salario = df_limpo.groupby('senioridade')['usd'].mean().sort_values(ascending=False).reset_index()

fig = px.bar(senioridade_media_salario,
             x='senioridade',
             y='usd',
             title='Média de Salário por Senioridade',
             labels={'senioridade': 'Nível de Senioridade', 'usd': 'Média de Salário Anual (USD)'},
             color='senioridade',
             )
fig.show()
```

```python
remoto_contagem = df_limpo['remoto'].value_counts().reset_index()
remoto_contagem.columns = ['tipo_trabalho', 'quantidade']

fig = px.pie(remoto_contagem,
             names='tipo_trabalho',
             values='quantidade',
             title='Proporção dos Tipos de Trabalho'
             )


fig.show()
```

```python
remoto_contagem = df_limpo['remoto'].value_counts().reset_index()
remoto_contagem.columns = ['tipo_trabalho', 'quantidade']

fig = px.pie(remoto_contagem,
             names='tipo_trabalho',
             values='quantidade',
             title='Proporção dos Tipos de Trabalho',
             hole=0.5
             )


fig.show()
```

```python
remoto_contagem = df_limpo['remoto'].value_counts().reset_index()
remoto_contagem.columns = ['tipo_trabalho', 'quantidade']

fig = px.pie(remoto_contagem,
             names='tipo_trabalho',
             values='quantidade',
             title='Proporção dos Tipos de Trabalho',
             hole=0.5
             )

fig.update_traces(textinfo='percent+label')
fig.show()
```
